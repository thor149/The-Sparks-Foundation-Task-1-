```python
#importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#reading the data and creating pandas dataframe
data = pd.read_csv('https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv')
df = pd.DataFrame(data)
df.head()
```

```python
#Creating a scatter plot
plt.scatter(df['Hours'], df['Scores'],marker='o')

#setting labels and title
plt.xlabel('Study Hours')
plt.ylabel('Scores')
plt.title('Study Hours vs Scores')

#showing the plot
plt.show()
```

```python
#attributes and outputs
x = df.iloc[:,:-1].values
y = df.iloc[:,1].values
```

```python
#splitting the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#creating and training the model over the train split
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
#using the trained model to predict the output on the test split
y_pred = model.predict(x_test)
y_pred
```

```python
#checking the accuracy of the model

#calculating themean squared error
mse = mean_squared_error(y_test, y_pred)

#calculating the root mean squared error
rmse = np.sqrt(mse)

#calculating the R-squared
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

```python
#using Scatter Plot to understand the relationship between actual and predicted value

plt.scatter(y_test, y_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Scatter Plot: Predicted vs. Actual')
plt.show()
```

```python
#using Line Plot to understand the relationship between actual and predicted value

indices = np.arange(len(y_test))

plt.plot(indices, y_test, label='Actual')
plt.plot(indices, y_pred, label='Predicted')
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Line Plot: Predicted vs. Actual')
plt.legend()
plt.show()
```
